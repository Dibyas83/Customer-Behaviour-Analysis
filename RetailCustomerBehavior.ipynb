we will find median review value of each category and fill

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median())) # find median of x's(category wise) values and fill in not available x's value
```

```python
df.isnull().sum()
```

make column names snake casing

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.info()
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

create a col to mark age groups

```python
labels = ['Young Adult','Adult', 'Middle-aged', 'Senior']  # created list of labels
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels) # asked to divide age to 4 category
```

```python
df[['age','age_group']].head(10)
```

```python
# convert text data of purchasing freq  into numdata column
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Quarterly': 90,
    'Monthly': 30,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90 
}
df['purchasing_freuency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchasing_freuency_days', 'frequency_of_purchases']].head(10)
```
